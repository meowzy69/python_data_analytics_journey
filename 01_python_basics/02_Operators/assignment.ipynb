# Assignment 2 - Online Shopping Billing System

## Objective

Use every operator you've learned to build a simple shopping bill.

```python
customer_name = "Ash"
product_name = "Smartphone"
price = 6000.0
quantity = 5
discount_percent = 15
wallet_balance = 50000.0

# Part 1 – Arithmetic Operators
# Calculate total price.
# Calculate discount amount.
# Calculate final bill.

total_price = price*quantity
dis_amt = total_price * (discount_percent/100)
final_bill = total_price - dis_amt

print("Total price =",total_price)
print("discount amount =", dis_amt)
print("Final bill =",final_bill)

# Part 2 – Comparison Operators

# Check:

# Is final bill greater than ₹5000?
# Is quantity at least 2?

print(final_bill > 5000)
print(quantity >= 2)

# Part 3 – Logical Operators

# Print whether the customer:

# Can afford the purchase
# Has enough wallet balance
# AND quantity is greater than 0

print("Affordable :",wallet_balance >= final_bill and quantity > 0)

# Part 4 – Assignment Operators

# Increase wallet balance by ₹1000.

# Subtract the bill.

# Print remaining balance.

wallet_balance += 1000.0
wallet_balance -= final_bill

print("new wallet balance",wallet_balance)

# Part 5 – Membership Operators
# create payment methods

payment_methods = ["UPI", "Card", "Cash"]

# Check whether:

# "UPI" is available
# "Cheque" is not available

print("UPI" in payment_methods)
print("Cheque" not in payment_methods)

# Part 6 – Identity Operators

cart = payment_methods

# print:
# payment_methods == cart
# paymenrt_methods is cart

print(payment_methods == cart)
print(payment_methods is cart)
```
